<span style="font-size:23px;">Complexity of Knowledge in Empirical Decision-Making</span><br>

<span style="font-weight:bold; font-size:17px;">Introduction</span><br>

Science is constructed upon empirically validated evidence that cumulatively forms our "knowledge." However, when we dissect the processes producing the data, matters of "truth" and "justice" become focal points, often blurred in real-world applications.

<span style="font-weight:bold; font-size:17px;">A Case Study: The CKD-EPI Equation and Cisplatin Therapy</span><br>

![Decision Making in Medicine](https://github.com/jhutrc/beta/blob/main/act_0/act_0_2/act_0_2_1/fig1_cisplatin.png?raw=true)

Consider a Urologist evaluating the impact of the new race-neutral CKD-EPI 2021 equation for determining eligibility for cisplatin therapy among bladder cancer patients. For over a decade, the standard has been the CKD-EPI 2009 equation, requiring four inputs:

<Details>
   <Summary>Definition of CKD-EPI</Summary>
   The CKD-EPI (Chronic Kidney Disease Epidemiology Collaboration) equation is a formula often used to estimate the Glomerular Filtration Rate (GFR), a measure of kidney function. The CKD-EPI 2009 equation, while more accurate than the older MDRD (Modification of Diet in Renal Disease) equation, raised issues of justice and equity in the United States due to its inclusion of race as a variable. It's worth noting that these issues have yet to garner attention in other countries.
</Details>

1. **Serum Creatinine (SCr)**: A byproduct of muscle metabolism, measured in mg/dL or ¬µmol/L.
  
2. **Age**: The patient's age in years.

3. **Sex**: Biological sex (Male or Female), relevant due to differences in muscle mass and consequently, creatinine levels.

4. **Race**: Originally included as a variable in the CKD-EPI 2009 equation, race has since become a point of ongoing debate due to ethical concerns.

<span style="font-weight:bold; font-size:17px;">The Original Equations</span><br>

The CKD-EPI 2009 equation relies on subgroup-specific formulations to calculate the Glomerular Filtration Rate (GFR).

- For females with SCr ‚â§ 0.7 mg/dL: 
   $GFR = 144 \times \left( \frac{{SCr}}{0.7} \right)^{-0.329} \times 0.993^{Age}$
  
- For females with SCr > 0.7 mg/dL: 
   $GFR = 144 \times \left( \frac{{SCr}}{0.7} \right)^{-1.209} \times 0.993^{Age}$

- For males with SCr ‚â§ 0.9 mg/dL: 
   $GFR = 141 \times \left( \frac{{SCr}}{0.9} \right)^{-0.411} \times 0.993^{Age}$

- For males with SCr > 0.9 mg/dL: 
   $GFR = 141 \times \left( \frac{{SCr}}{0.9} \right)^{-1.209} \times 0.993^{Age}$

- For African Americans, the estimated GFR is then multiplied by 1.159.

<span style="font-weight:bold; font-size:17px;">The Impact of a Race-Free Equation</span><br>

Adopting the new race-neutral equation opens a plethora of questions that warrant rigorous empirical scrutiny:

- What is the new classification of bladder cancer patients?
- How does this impact eligibility for cisplatin therapy?
- Are there expected changes in adverse effects?

Such questions require methodical empirical investigation, inviting the reader to consider possible study designs.
